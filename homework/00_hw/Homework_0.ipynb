## Import libraries

```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
```

## Load a grayscale image

```python
# read in the image
img = cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'), cv2.IMREAD_GRAYSCALE)
```

## View the image

```python
# render the image
plt.imshow(img, cmap='gray')
plt.title('Apollo 11 Launch')
plt.show()
```

## Crop the image

```python
# crop the image
img_cropped = img[100:700, 400:800]

# render the image
plt.imshow(img_cropped, cmap='gray')
plt.title('Apollo 11 Launch Cropped')
plt.show()
```

## Edit the brightness

```python
matrix = np.ones(img.shape, dtype="uint8") * 50
img_bright = cv2.add(img, matrix)
plt.imshow(img_bright, cmap="gray"); plt.title('Apollo 11 Launch Brighter')
plt.show()
```

## Rotate the image

```python
img_flipped = cv2.flip(img, 0)
plt.imshow(img_flipped, cmap='gray'); plt.title('Apollo 11 Launch Rotated')
plt.show()
```

## Save the image

```python
cv2.imwrite(os.path.relpath('Data/apollo_11_mod.jpg'), img_flipped)
```
